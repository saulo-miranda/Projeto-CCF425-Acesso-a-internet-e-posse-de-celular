# Dataframe da Idade por Região

### Leitura separada pelo ano

```python
dfs_idade_regiao = []
def ler(arquivos, dfs):
    for arq in arquivos:
        dfIdade = pd.read_excel(arq, sheet_name="Tab 1.1.2.1 a 1.1.2.3", usecols="A:G", header=4)
        dfs.append(dfIdade)

ler(arquivos, dfs_idade_regiao)
```

### Tratamento dos dados



Ao exibir pode-se notar, tabelas não nomeadas, assim como uma nota de fonte, e valores nulos, além de também os valores somados respectivos a cada sexo, que não é interessante manter para o estudo do dataframe, portanto deve-se tratar tudo isso.

```python
dfs_idade_regiao[1]
```

```python
def tratar(dfs):
    for i in range(len(dfs)):
        dfs[i] = dfs[i].drop(dfs[i].index[[0, 13, 26, 39, 40]])
        dfs[i].rename(columns={'Unnamed: 0': 'Grupo de Idades', 'Unnamed: 1': 'Brasil'}, inplace=True)
tratar(dfs_idade_regiao)
dfs_idade_regiao[0]
```

### Separação dos dados


Podemos notar que os dados, de Homens e Mulheres assim como seus totais estão juntos, portanto é interessante separar cada um deles criando uma coluna por sexo.

```python
def separar(dfs):
    newdfs = []
    for df in dfs:
        s = pd.Series([])
        dfIdadeHomen = df.iloc[12:24]  # Homens
        dfIdadeMulher = df.iloc[24:]  # Mulheres
        dfIdadeHomen.insert(7, "Sexo", "Masculino")
        dfIdadeMulher.insert(7, "Sexo", "Feminino")
        dfIdadeTotal = pd.concat([dfIdadeMulher, dfIdadeHomen], sort=False)
        newdfs.append(dfIdadeTotal)
    return newdfs
  
dfs_idade_regiao = separar(dfs_idade_regiao)
dfs_idade_regiao[0] #Dataframe das idades de homens por região de 2013 que utilizam a internet
```

#### Utilizando Regressão para prever os dados em anos futuros


```python
from traitlets.traitlets import default
from sklearn.linear_model import LinearRegression


#Remodelando os dados para regressão

temp = newdf.copy()
temp['Grupo de Idades'] = newdf['Grupo de Idades'].astype('category')
#temp['Grupo de Idades'] = temp['Grupo de Idades'].cat.codes
temp['Sexo'] = newdf['Sexo'].astype('category')
#temp['Sexo'] = temp['Sexo'].cat.codes
temp = pd.get_dummies(temp)


def retornaGrupos(x):
  lista = {
      0: "Grupo de Idades_10 a 14 anos",
      1: "Grupo de Idades_15 a 17 anos",
      2: "Grupo de Idades_18 ou 19 anos",
      3: "Grupo de Idades_20 a 24 anos",
      4: "Grupo de Idades_25 a 29 anos",
      5: "Grupo de Idades_30 a 34 anos",
      6: "Grupo de Idades_35 a 39 anos",
      7: "Grupo de Idades_40 a 44 anos",
      8: "Grupo de Idades_45 a 49 anos",
      9: "Grupo de Idades_50 a 54 anos",
      10: "Grupo de Idades_55 a 59 anos",
      11: "Grupo de Idades_60 anos ou mais",
  }
  return lista.get(x,default)


def retornaSexo(x):
  lista = {
      0: 'Sexo_Feminino',
      1: 'Sexo_Masculino'
  }
  return lista.get(x,default)


#Renomeando as colunas do df

def renomearGrupos(x):
  lista = {
      0: "10 a 14 anos",
      1: "15 a 17 anos",
      2: "18 ou 19 anos",
      3: "20 a 24 anos",
      4: "25 a 29 anos",
      5: "30 a 34 anos",
      6: "35 a 39 anos",
      7: "40 a 44 anos",
      8: "45 a 49 anos",
      9: "50 a 54 anos",
      10: "55 a 59 anos",
      11: "60 anos ou mais"
  }
  return lista.get(x,default)

def renomearSexo(x):
  lista = {
      0: "Feminino",
      1: "Masculino",
    }
  return lista.get(x,default)

#Calcular dado a partir de seus coeficientes.

def calcula(ano,coef1,grupo,coef2,sexo,coef3,interc):
  return interc + ano * coef1 + grupo *coef2 + sexo * coef3

#Aplicando regressão linear 

def aplicarRegressao(lscore,df,ldf,col,ano,grupo,sexo):
  X = df[['Ano',retornaGrupos(grupo),retornaSexo(sexo)]]
  Y = df[col]
  lm = LinearRegression()
  lm.fit(X, Y)
  lscore.append(lm.score(X,Y))
  return calcula(ano,lm.coef_[0],1,lm.coef_[1],1,lm.coef_[2],lm.intercept_)

#Utilização de regressão para criação de um novo dataframe.

def regressao(df,ano,cols,mscore):
  ldf = []
  data = []
  lscore = []
  for i in range(2):
    for j in range(12):
      ldf.append(j)  
      for col in cols :
        ldf.append(aplicarRegressao(lscore,df,ldf,col,ano,j,i))
      ldf.append(i)
      ldf.append(ano)
      data.append(ldf)
      mscore.append(lscore)
      lscore = []
      ldf = []
  rdf = pd.DataFrame(data, columns=['Grupo de Idades','Brasil','Norte','Nordeste','Sudeste','Sul','Centro-Oeste','Sexo','Ano'])    
  rdf['Grupo de Idades'] = rdf['Grupo de Idades'].apply(lambda x: renomearGrupos(x))
  rdf['Sexo'] = rdf['Sexo'].apply(lambda x: renomearSexo(x))  
  return rdf

#Obtendo Dataframe a partir da regressão para o ano de 2016 e 2017.
mscore = []
cols = ['Brasil','Norte','Nordeste','Sudeste','Sul','Centro-Oeste']
df2016 = regressao(temp,2016,cols,mscore)
df2017 = regressao(temp,2017,cols,mscore)
#Exibir o R2 (score) para as respectivas tabelas geradas no novo DF.
mscore

```

Como visto acima estão os coeficientes de determinação (R2) para cada grupo de dados , os coeficientes não estão tão elevados por possuir poucas categorias de anos, e pela variavel ano em si não ser o único fator determinante, entretando ainda é possivel fazer alguma previsão com uma margem de erro, como faremos a seguir.

Utilizando Regressão para prever os dados em anos futuros

```python
from traitlets.traitlets import default
from sklearn.linear_model import LinearRegression

#Remodelando os dados para regressão

temp = newdf.copy()
temp['Sexo e nivel de instrucao'] = newdf['Sexo e nivel de instrucao'].astype('category')
temp['Sexo'] = newdf['Sexo'].astype('category')
temp = pd.get_dummies(temp)


def retornaGrupos(x):
  lista = {
      0: "Sexo e nivel de instrucao_Sem instrução",
      1: "Sexo e nivel de instrucao_Fundamental incompleto",
      2: "Sexo e nivel de instrucao_Fundamental completo",
      3: "Sexo e nivel de instrucao_Médio incompleto",
      4: "Sexo e nivel de instrucao_Médio completo",
      5: "Sexo e nivel de instrucao_Superior incompleto",
      6: "Sexo e nivel de instrucao_Superior completo",
      7: "Sexo e nivel de instrucao_Não determinado",
  }
  return lista.get(x,default)


def retornaSexo(x):
  lista = {
      0: 'Sexo_Feminino',
      1: 'Sexo_Masculino'
  }
  return lista.get(x,default)


#Renomeando as colunas do df

def renomearGrupos(x):
  lista = {
      0: "Sem instrução",	
      1: "Fundamental incompleto",
      2: "Fundamental completo", 
      3: "Médio incompleto",
      4: "Médio completo",
      5: "Superior incompleto",
      6: "Superior completo",	
      7: "Não determinado",
  }
  return lista.get(x,default)

def renomearSexo(x):
  lista = {
      0: "Feminino",
      1: "Masculino",
    }
  return lista.get(x,default)

#Calcular dado a partir de seus coeficientes.

def calcula(ano,coef1,grupo,coef2,sexo,coef3,interc):
  return interc + ano * coef1 + grupo *coef2 + sexo * coef3

#Aplicando regressão linear 

def aplicarRegressao(df,ldf,col,ano,grupo,sexo):
  X = df[['Ano',retornaGrupos(grupo),retornaSexo(sexo)]]
  Y = df[col]
  lm = LinearRegression()
  lm.fit(X, Y)
  #print("grupo ",grupo," sexo",sexo, "coef ", lm.coef_)
  return calcula(ano,lm.coef_[0],1,lm.coef_[1],1,lm.coef_[2],lm.intercept_)

#Utilização de regressão para criação de um novo dataframe.

def regressao(df,ano,cols):
  ldf = []
  data = []
  for i in range(2):
    for j in range(8):
      ldf.append(j)  
      for col in cols :
        ldf.append(aplicarRegressao(df,ldf,col,ano,j,i))
      ldf.append(i)
      ldf.append(ano)
      data.append(ldf)
      ldf = []
  rdf = pd.DataFrame(data, columns=['Sexo e nivel de instrucao','Brasil','Norte','Nordeste','Sudeste','Sul','Centro-Oeste','Sexo','Ano'])    
  rdf['Sexo e nivel de instrucao'] = rdf['Sexo e nivel de instrucao'].apply(lambda x: renomearGrupos(x))
  rdf['Sexo'] = rdf['Sexo'].apply(lambda x: renomearSexo(x))  
  return rdf

#Obtendo Dataframe a partir da regressão para o ano de 2016 e 2017.

cols = ['Brasil','Norte','Nordeste','Sudeste','Sul','Centro-Oeste']
df2016 = regressao(temp,2016,cols)
df2017 = regressao(temp,2017,cols)
```

### Funções para filtrar e separar dataframes

```python
def filtrar_tabelas(arquivos, tabela, prefixo, linhas, nome_coluna):
  anos = ['2015', '2014', '2013']
  for f in arquivos:
    df = pd.read_excel(f, sheet_name=tabela, usecols='A:G')
    df.columns = [nome_coluna, "Brasil", "Norte","Nordeste","Sudeste","Sul","Centro-Oeste"]
    df2 = df
    df = df.drop(df.index[[0,1,2,3]])
    df = df[df.index < linhas]
    df.to_csv('/content/'+prefixo+anos.pop()+'.csv')
```

```python
def separar_dataframes(csvs, limites):
  dfs = []
  for c in csvs:  
    df = pd.read_csv(c)
    df = df.drop(columns=['Unnamed: 0'])
    df_u = df.iloc[limites[1]:limites[2]:]
    df_n = df.iloc[limites[2]:,:]
    df_u.insert(7, 'Utilizaram', 'Sim')
    df_n.insert(7, 'Utilizaram', 'Nao')
    df_total = pd.concat([df_u, df_n], sort=False)
    df_total = df_total.drop(df_total.index[[limites[0],limites[1]]])
    dfs.append(df_total)
  return dfs
```

### Função para recuperar profissoes que mais usaram internet por região

```python
def mais_usaram_regiao(dfs, regioes):
  retorno = []
  for d in dfs:
    for r in regioes:
      i = d[d['Utilizaram'] == 'Sim']
      i = i[i[r]==i[r].max()].index[0]
      retorno.append(d['Profissao'][i])
  return retorno
```

```python
from pandas._libs.lib import array_equivalent_object
newdf1 = dfs_idade_regiao[0].copy()
newdf2 = dfs_idade_regiao[1].copy()
newdf3 = dfs_idade_regiao[2].copy()
newdf1.insert(8,"Ano","2013")
newdf2.insert(8,"Ano","2014")
newdf3.insert(8,"Ano","2015")
newdf = pd.concat([newdf1,newdf2,newdf3], sort=False)
groupIdades = newdf.groupby('Ano').sum().reset_index()

regioes = ['Centro-Oeste', 'Nordeste', 'Norte', 'Sudeste', 'Sul']
anos = ["2013", "2014", "2015"]
def filtro_regiao_ano(v_regiao, v_ano, group):
  arr_anos = []
  arr_usuarios = []
  group.set_index('Ano', inplace=True)
  for a in v_ano:
    for r in v_regiao:
      arr_usuarios.append(group.at[a,r])
      arr_anos.append(a)
  return arr_anos, arr_usuarios

v_anos, users = filtro_regiao_ano(regioes, anos, groupIdades)

df_regioes = pd.DataFrame()
df_regioes['Regioes'] = regioes + regioes + regioes
df_regioes['Usuarios'] = users
df_regioes['Anos'] = v_anos
df_regioes = df_regioes.sort_values(by='Usuarios')
plt.figure(figsize=(10,10))
g=sns.barplot(data=df_regioes, y='Regioes', x='Usuarios',palette="Set1",hue= "Anos" ,orient='h')
g.set_ylabel('Região')
g.set_xlabel('Usuários')
g.set_title('Usuários de internet no Brasil por Região (2013-2015)')
plt.show()
```
